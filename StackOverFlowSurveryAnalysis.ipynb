### What are the distinct professions reported in surveys?

```python
profession = data['Professional'].value_counts()
professional = list(profession)
colors = ['gold', 'yellowgreen', 'lightcoral', 'lightskyblue', 'black']
labels = ['Professional\ndeveloper', 'Student', 'Professional\nnon-developer\n who\nsometimes\nwrites code', 'Used to\nbe a\nprofessional\ndeveloper', 'None of these']
plt.title('Which of the following best describes you?')
plt.bar(labels, professional,width=0.5, align='center', alpha=0.5, color=colors)
plt.show()
profession_developer_count = professional[0]
```

### Let's see if programming can be a hobby or just a way to earn handsome amount

```python
programHobby = data['ProgramHobby'].value_counts()
#Code to draw plot
labels = ['Yes, I program as a hobby','Yes, both', 'No', 'Yes, I contribute to open source projects']
programHobby = list(programHobby)
colors = ['gold', 'yellowgreen', 'lightcoral', 'lightskyblue']
explode = (0.05, 0, 0, 0.3)  # explode 1st slice

# Plot
plt.pie(programHobby, explode=explode, labels=labels, colors=colors,
autopct=lambda p: '{:.0f}'.format((p/100)*sum(programHobby)), shadow=True, startangle=140)
plt.axis('equal')
plt.show()
```

Sounds interesting. Right! So, a large number of people who are actively visiting stack overflow program because it excites them.
